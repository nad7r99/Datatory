```python
df = df.drop_duplicates()
```

### Standardize_Name(Consistency)

```python
df['Status'].unique()
```

```python
df["Status"] = df["Status"].str.replace("UNEMPLOYED", "Unemployed")

df["Status"] = df["Status"].str.replace("EMPLOYED", 'employed').str.title()
```

```python
df["Status"].unique()
```

```python
df["Age Group"].unique()
```

```python
df["Age Group"] = df["Age Group"].str.replace("25_34", "25-34")

df["Age Group"] = df["Age Group"].str.replace("18 to 24",'18-24').str.replace('35 - 44', '35-44')
```

```python
df["Age Group"].unique()
```

```python
df["Education"].unique()
```

```python
df["Education"] = df["Education"].str.replace("high school", 'High School')

df["Education"] = df["Education"].str.replace("Masters", "Master")

df["Education"] = df["Education"].str.replace('Ph.D', 'PhD').str.replace('Bachelors', "Bachelor's")
```

```python
df["Education"].unique()
```

```python
df["Industry"].unique()
```

```python
df["Industry"] = df["Industry"].str.replace("Tech ", "Technology").str.replace("Healthcare", "Health")
```

```python
df["Industry"].unique()
```

```python
df["Location"].unique()
```

```python
df["Location"] = df["Location"].str.title()
```

```python
df["Location"].unique()
```

```python
df['AI Risk'].unique()
```

```python
df['AI Risk'] = df['AI Risk'].str.replace("moderate", "Medium").str.lower().str.title()
```

```python
df["AI Risk"].unique()
```

```python
df['Years of Experience'].unique()
```

```python
df["Monthly Salary (INR)"].unique()
```

```python
df["Date Recorded"].unique()
```

### Handling_Nulls_ETL(Completeness)

```python
df["Status"].notnull().value_counts()
```

```python
df["Status"]=df["Status"].fillna("Unknown")
```

```python
df["Status"].isna().value_counts()
```

```python
df["Age Group"].notnull().value_counts()
```

```python
df['Age Group']=df["Age Group"].fillna('0')
```

```python
df["Age Group"].isna().value_counts()
```

```python
df["Education"].isna().value_counts()
```

```python
df["Education"]=df["Education"].fillna("No Info")
```

```python
df["Education"].isna().value_counts()
```

```python
df["Industry"].isna().value_counts()
```

```python
df["Industry"]=df['Industry'].fillna("No Data")
```

```python
df['Industry'].isna().value_counts()
df["Industry"].unique()
```

```python
df['Location'].isna().value_counts()
```

```python
df["Location"]=df['Location'].fillna(method="ffill")
```

```python
df['Location'].isna().value_counts()
```

```python
df["AI Risk"].isna().value_counts()
```

```python
df["AI Risk"]=df["AI Risk"].fillna("No Level")
```

```python
df["AI Risk"].isna().value_counts()
```

```python
df['Years of Experience'].isna().value_counts()
```

```python
df["Years of Experience"] = df["Years of Experience"].fillna(df["Years of Experience"].mean()).astype(int)
```

```python
df['Years of Experience'].isna().value_counts()

# df["Years of Experience"].unique()
```

```python
df["Monthly Salary (INR)"].isna().value_counts()
```

```python
df["Monthly Salary (INR)"]=df["Monthly Salary (INR)"].fillna(df["Monthly Salary (INR)"].mean()).astype(int)
```

```python
df["Monthly Salary (INR)"].isna().value_counts()
```

```python
df['Date Recorded'].isna().value_counts()
```
